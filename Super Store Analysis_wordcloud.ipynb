```python
import pandas as pd
import os
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
os.chdir("C:/Users/LENOVO/Documents/Data Analytics/Side Hustle data")
```

```python
data = pd.ExcelFile("Super+Store+Data.xlsx")
data.sheet_names
```

```python
df = data.parse("Orders")
df.head()
```

```python
sub_category = df.groupby("Sub-Category").agg({"Quantity":"sum"})
sub_cat_sorted = sub_category.sort_values(by="Quantity")
sub_cat_sorted
```

```python
type(sub_category)
```

```python
sub_category_count = sub_cat_sorted["Quantity"].to_dict()
```

```python
#Generating the string object to be used in the wordcloud
word_list =[]
for key, value in sub_category_count.items():
    for i in range(value):        # Adding the sub_category name to the list by iterating through the quantity value 
        word_list.append(key)
#converting it into a string
word_text = " ".join(word_list)
```

```python
#Creating Super Store wordcloud
wordcloud = WordCloud(collocations = False, background_color= "white", 
                      color_func=lambda *args, **kwargs: (32,56,140)).generate(word_text)
wordcloud.recolor()
fig, ax = plt.subplots()
plt.imshow(wordcloud, interpolation = "bilinear")
plt.axis("off")
plt.show()
```
